```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the Boston House Price Dataset

```python
house_price_dataset = pd.read_csv(r"C:\Users\Vikram\OneDrive\Desktop\ML Projects\Projects\3.1. Boston House Price Dataset.csv")
```

```python
house_price_dataset.head()
```

```python
# Checking the number of rows and columns in the DataFrame
house_price_dataset.shape
```

```python
# Check for missing values
house_price_dataset.isnull().sum()
```

```python
# Statistical measures of the Dataset
house_price_dataset.describe()
```

Understanding the correlation between various features in the Dataset

1. Positive Correlation

2. Negative Correlation

```python
correlation = house_price_dataset.corr()
```

```python
# Constructing a Heatmap to understand the correlation
plt.figure(figsize=(10,10))
plt.title('Correlation between various features in the Dataset')
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Splitting the Data and Target

```python
X = house_price_dataset.drop(columns='MEDV')
Y = house_price_dataset['MEDV']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Data into Training data and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor

```python
# Loading the model
model = XGBRegressor()
```

```python
# Training the model with X_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on Training Data

```python
# Accuracy for prediction on Training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

Errors for Training data

```python
# R-squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R-squared error :", score_1)  # Lesser the value, more accurate is the model
print("Mean Absolute error :", score_2)
```

As the errors are less, our model is performing well.

Visualizing the Actual Prices and Predicted Prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs. Predicted Price")
plt.show()
```

Prediction on Test data

```python
# Accuracy for prediction on Test data
test_data_prediction = model.predict(X_test)
```

```python
# R-squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R-squared error :", score_1)  # Lesser the value, more accurate is the model
print("Mean Absolute error :", score_2)
```
